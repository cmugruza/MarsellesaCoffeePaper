This program/script is to reproduce the figure 2 in the article "Evaluation of the phenotypic performance of the coffee variety MarsellesaÂ® over 1400 meters a.s.l as an alternative for coffee growing": 

The figure accounts on the Vegetative behavior of coffee varieties according to plant height (A: top left), number of branches (B: top right), branch length (C: center), and number of nodes per branch (D: down). Black: Block 1, Gray: Block 2, Light gray: Block 3.

Figure 2(A): Top left, number of branches.

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()

y = [109.86,	107.37,	90.58,	99.05,	100.47,	88.66,	90.41,	100.17,	81.42] 
y_std_dev = [20.50,	9.07,	8.52,	11.45,	6.93,	10.93,	10.81,	12.505,	8.26]
x=np.arange(len(y))
plt.figure(figsize=(6, 6)) 
#define chart 
fig, ax = plt.subplots()

#create chart
ax.bar(x=x, #x-coordinates of bars
       height=y, #height of bars
       yerr=y_std_dev, #error bar width
       capsize=4) #length of error bar caps

barlist=plt.bar(np.arange(len(y)), y, edgecolor="black", linewidth=3)#, width = .5)
for i in [0,3,6]:
  barlist[i].set_color('#2F2F2F')
  barlist[i+1].set_color('#888888')
  barlist[i+2].set_color('#E0E0E0')
  
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = [ '1', '2', '3', '1', '2', '3', '1', '2', '3']#'Testing'
plt.xticks(x, labels)#, rotation='vertical')
ax.set_xticklabels(labels)

plt.rc('xtick', labelsize=18) 
plt.rc('ytick', labelsize=18) 
plt.xlabel('Red Caturra     Catimor      Marsellesa  ', fontsize=18)#\n\n''Variety', fontsize=18)  #previous 16
plt.ylabel('Plant height', fontsize=24)       
plt.savefig('Fig2a.png', dpi = 300, bbox_inches='tight', pad_inches=0.0)
#plt.errorbar(x, y, yerr=yerr, label='both limits (default)')
#plt.legend(loc='lower right')
```
